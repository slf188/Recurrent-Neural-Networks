## What is the Vanishing Gradient?
Issues can arise during backpropagation in deep RNNs:
* Vanishing gradients: As we go back to the previous layers, gradients often get smaller, causing weights no never change.
* Exploding gradients: On the way back gradients explode, causing issues.

This happens either when the slopes or derivatives get big.

Choosing different initialization of weights can help alleviate these issues:
* Xavier Initialization
* Gradient Clipping: Gradients are cut off before reaching a limit

## What is LSTMS and Gru?
Another issue is, the RNN will begin to forget the first inputs, as information is lost at each step, so we need a long-term memory.

The Long Short-Term Memory(LSTM) cell can hel address this issue.
In LSTM we have 4 important components:
* Forget Gate: What to forget from the previous memory units
* Input Gate: What to accept into the neuron
* Update Gate: Update the memories
* Output Gate: Output the new long-term memories
